```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

## Economic measures

### Antecedent

```python
manifacturing_cost=ctrl.Antecedent(np.arange(0,11,1), 'Manifacturing cost')
quality=ctrl.Antecedent(np.arange(0,11,1), 'Quality')

responsiveness=ctrl.Antecedent(np.arange(0,11,1), 'Responsiveness')
flexibility=ctrl.Antecedent(np.arange(0,11,1), 'Flexibility')

fis1_eco_out = ctrl.Antecedent(np.arange(0,11,1), 'Fis1 eco out')
fis2_eco_out = ctrl.Antecedent(np.arange(0,11,1), 'Fis2 eco out')
```

### Membership functions

```python
manifacturing_cost.automf(3)
quality.automf(3)

responsiveness.automf(3)
flexibility.automf(3)

fis1_eco_out.automf(3)
fis2_eco_out.automf(3)
```

### Consequent

```python
#FIS1_ECO
fis1_eco = ctrl.Consequent(np.arange(0,11,1),'FIS1_ECO')
fis1_eco['poor'] = fuzz.trimf(fis1_eco.universe, [0,0,5])
fis1_eco['average'] = fuzz.trimf(fis1_eco.universe, [0,5,10])
fis1_eco['good'] = fuzz.trimf(fis1_eco.universe, [5,10,10])

#FIS2_ECO
fis2_eco = ctrl.Consequent(np.arange(0,11,1),'FIS2_ECO')
fis2_eco['poor'] = fuzz.trimf(fis2_eco.universe, [0,0,5])
fis2_eco['average'] = fuzz.trimf(fis2_eco.universe, [0,5,10])
fis2_eco['good'] = fuzz.trimf(fis2_eco.universe, [5,10,10])

#FIS3_ECO
fis3_eco = ctrl.Consequent(np.arange(0,11,1),'FIS3_ECO')
fis3_eco['poor'] = fuzz.trimf(fis3_eco.universe, [0,0,5])
fis3_eco['average'] = fuzz.trimf(fis3_eco.universe, [0,5,10])
fis3_eco['good'] = fuzz.trimf(fis3_eco.universe, [5,10,10])
```

```python
manifacturing_cost.view()
```

```python
fis1_eco.view()
```

### Rules

```python
#FIS1_ECO
rule1fis1_eco = ctrl.Rule(manifacturing_cost['poor'] & quality['poor'], fis1_eco['poor'])
rule2fis1_eco = ctrl.Rule(manifacturing_cost['average'] & quality['poor'], fis1_eco['poor'])
rule3fis1_eco = ctrl.Rule(manifacturing_cost['good'] & quality['average'], fis1_eco['average'])
rule4fis1_eco = ctrl.Rule(manifacturing_cost['poor'] & quality['average'], fis1_eco['poor'])
rule5fis1_eco = ctrl.Rule(manifacturing_cost['average'] & quality['average'], fis1_eco['average'])
rule6fis1_eco = ctrl.Rule(manifacturing_cost['good'] & quality['average'], fis1_eco['average'])
rule7fis1_eco = ctrl.Rule(manifacturing_cost['poor'] & quality['good'], fis1_eco['average'])
rule8fis1_eco = ctrl.Rule(manifacturing_cost['average'] & quality['good'], fis1_eco['average'])
rule9fis1_eco = ctrl.Rule(manifacturing_cost['good'] & quality['good'], fis1_eco['good'])

#FIS2_ECO
rule1fis2_eco = ctrl.Rule(responsiveness['poor'] & flexibility['poor'], fis2_eco['poor'])
rule2fis2_eco = ctrl.Rule(responsiveness['average'] & flexibility['poor'], fis2_eco['poor'])
rule3fis2_eco = ctrl.Rule(responsiveness['good'] & flexibility['average'], fis2_eco['average'])
rule4fis2_eco = ctrl.Rule(responsiveness['poor'] & flexibility['average'], fis2_eco['poor'])
rule5fis2_eco = ctrl.Rule(responsiveness['average'] & flexibility['average'], fis2_eco['average'])
rule6fis2_eco = ctrl.Rule(responsiveness['good'] & flexibility['average'], fis2_eco['average'])
rule7fis2_eco = ctrl.Rule(responsiveness['poor'] & flexibility['good'], fis2_eco['average'])
rule8fis2_eco = ctrl.Rule(responsiveness['average'] & flexibility['good'], fis2_eco['average'])
rule9fis2_eco = ctrl.Rule(responsiveness['good'] & flexibility['good'], fis2_eco['good'])
```

```python
#FIS3_ECO
rule1fis3_eco = ctrl.Rule(fis1_eco_out['poor'] & fis2_eco_out['poor'], fis3_eco['poor'])
rule2fis3_eco = ctrl.Rule(fis1_eco_out['average'] & fis2_eco_out['poor'], fis3_eco['poor'])
rule3fis3_eco = ctrl.Rule(fis1_eco_out['good'] & fis2_eco_out['average'], fis3_eco['average'])
rule4fis3_eco = ctrl.Rule(fis1_eco_out['poor'] & fis2_eco_out['average'], fis3_eco['poor'])
rule5fis3_eco = ctrl.Rule(fis1_eco_out['average'] & fis2_eco_out['average'], fis3_eco['average'])
rule6fis3_eco = ctrl.Rule(fis1_eco_out['good'] & fis2_eco_out['average'], fis3_eco['average'])
rule7fis3_eco = ctrl.Rule(fis1_eco_out['poor'] & fis2_eco_out['good'], fis3_eco['average'])
rule8fis3_eco = ctrl.Rule(fis1_eco_out['average'] & fis2_eco_out['good'], fis3_eco['average'])
rule9fis3_eco = ctrl.Rule(fis1_eco_out['good'] & fis2_eco_out['good'], fis3_eco['good'])
```

```python
rule1fis3_eco.view()
```

```python
rule1fis2_eco.view()
```

```python
fis1_eco_ctrl = ctrl.ControlSystem([rule1fis1_eco,rule2fis1_eco,rule3fis1_eco,rule4fis1_eco,rule5fis1_eco,rule6fis1_eco,rule7fis1_eco,rule8fis1_eco,rule9fis1_eco])
fis2_eco_ctrl = ctrl.ControlSystem([rule1fis2_eco,rule2fis2_eco,rule3fis2_eco,rule4fis2_eco,rule5fis2_eco,rule6fis2_eco,rule7fis2_eco,rule8fis2_eco,rule9fis2_eco])
fis3_eco_ctrl = ctrl.ControlSystem([rule1fis3_eco,rule2fis3_eco,rule3fis3_eco,rule4fis3_eco,rule5fis3_eco,rule6fis3_eco,rule7fis3_eco,rule8fis3_eco,rule9fis3_eco])
```

```python
fissing1_eco = ctrl.ControlSystemSimulation(fis1_eco_ctrl)
fissing2_eco = ctrl.ControlSystemSimulation(fis2_eco_ctrl)
fissing3_eco = ctrl.ControlSystemSimulation(fis3_eco_ctrl)
```

```python
#Input
fissing1_eco.input['Quality'] = 2
fissing1_eco.input['Manifacturing cost'] = 2
fissing2_eco.input['Responsiveness'] = 8
fissing2_eco.input['Flexibility'] = 8
#Crunch the numbers
fissing1_eco.compute()
fissing2_eco.compute()
fissing3_eco.input['Fis1 eco out'] = fissing1_eco.output['FIS1_ECO']
fissing3_eco.input['Fis2 eco out'] = fissing2_eco.output['FIS2_ECO']
fissing3_eco.compute()
```

```python
print(fissing1_eco.output['FIS1_ECO'])
fis1_eco.view(sim=fissing1_eco)
```

```python
print(fissing2_eco.output['FIS2_ECO'])
fis2_eco.view(sim=fissing2_eco)
```

## Environmental measures

### Antecedent

```python
material_usage=ctrl.Antecedent(np.arange(0,10,1), 'Material usage')
energy_usage=ctrl.Antecedent(np.arange(0,10,1), 'Energy usage')

water_usage=ctrl.Antecedent(np.arange(0,10,1), 'Water usage')
waste=ctrl.Antecedent(np.arange(0,10,1), 'Waste')

emissions=ctrl.Antecedent(np.arange(0,10,1), 'Emissions')

fis1_env_out = ctrl.Antecedent(np.arange(0,10,1), 'Fist 1 env out')
fis2_env_out = ctrl.Antecedent(np.arange(0,10,1), 'Fist 2 env out')
fis3_env_out = ctrl.Antecedent(np.arange(0,10,1), 'Fist 3 env out')
```

### Membership functions

```python
material_usage.automf(3)
energy_usage.automf(3)

water_usage.automf(3)
waste.automf(3)

emissions.automf(3)

fis1_env_out.automf(3)
fis2_env_out.automf(3)
fis3_env_out.automf(3)
```

### Consequent

```python
#FIS1_ENV
fis1_env= ctrl.Consequent(np.arange(0,11,1),'FIS1_ENV')
fis1_env['poor'] = fuzz.trimf(fis1_env.universe, [0,0,5])
fis1_env['average'] = fuzz.trimf(fis1_env.universe, [0,5,10])
fis1_env['good'] = fuzz.trimf(fis1_env.universe, [5,10,10])

#FIS2_ENV
fis2_env = ctrl.Consequent(np.arange(0,11,1),'FIS2_ENV')
fis2_env['poor'] = fuzz.trimf(fis2_env.universe, [0,0,5])
fis2_env['average'] = fuzz.trimf(fis2_env.universe, [0,5,10])
fis2_env['good'] = fuzz.trimf(fis2_env.universe, [5,10,10])

#FIS3_ENV
fis3_env = ctrl.Consequent(np.arange(0,11,1),'FIS3_ENV')
fis3_env['poor'] = fuzz.trimf(fis3_env.universe, [0,0,5])
fis3_env['average'] = fuzz.trimf(fis3_env.universe, [0,5,10])
fis3_env['good'] = fuzz.trimf(fis3_env.universe, [5,10,10])

#FIS4_ENV
fis4_env = ctrl.Consequent(np.arange(0,11,1),'FIS4_ENV')
fis4_env['poor'] = fuzz.trimf(fis4_env.universe, [0,0,5])
fis4_env['average'] = fuzz.trimf(fis4_env.universe, [0,5,10])
fis4_env['good'] = fuzz.trimf(fis4_env.universe, [5,10,10])
```

### Rules

```python
#FIS1_ENV
rule1fis1_env = ctrl.Rule(material_usage['poor'] & energy_usage['poor'], fis1_env['poor'])
rule2fis1_env = ctrl.Rule(material_usage['average'] & energy_usage['poor'], fis1_env['poor'])
rule3fis1_env = ctrl.Rule(material_usage['good'] & energy_usage['average'], fis1_env['average'])
rule4fis1_env = ctrl.Rule(material_usage['poor'] & energy_usage['average'], fis1_env['poor'])
rule5fis1_env = ctrl.Rule(material_usage['average'] & energy_usage['average'], fis1_env['average'])
rule6fis1_env = ctrl.Rule(material_usage['good'] & energy_usage['average'], fis1_env['average'])
rule7fis1_env = ctrl.Rule(material_usage['poor'] & energy_usage['good'], fis1_env['average'])
rule8fis1_env = ctrl.Rule(material_usage['average'] & energy_usage['good'], fis1_env['average'])
rule9fis1_env = ctrl.Rule(material_usage['good'] & energy_usage['good'], fis1_env['good'])

#FIS2_ENV
rule1fis2_env = ctrl.Rule(water_usage['poor'] & waste['poor'], fis2_env['poor'])
rule2fis2_env = ctrl.Rule(water_usage['average'] & waste['poor'], fis2_env['poor'])
rule3fis2_env = ctrl.Rule(water_usage['good'] & waste['average'], fis2_env['average'])
rule4fis2_env = ctrl.Rule(water_usage['poor'] & waste['average'], fis2_env['poor'])
rule5fis2_env = ctrl.Rule(water_usage['average'] & waste['average'], fis2_env['average'])
rule6fis2_env = ctrl.Rule(water_usage['good'] & waste['average'], fis2_env['average'])
rule7fis2_env = ctrl.Rule(water_usage['poor'] & waste['good'], fis2_env['average'])
rule8fis2_env = ctrl.Rule(water_usage['average'] & waste['good'], fis2_env['average'])
rule9fis2_env = ctrl.Rule(water_usage['good'] & waste['good'], fis2_env['good'])

#FIS3_ENV = FIS2_ENV RESULT
rule1fis3_env = ctrl.Rule(fis2_env_out['poor'] & emissions['poor'], fis3_env['poor'])
rule2fis3_env = ctrl.Rule(fis2_env_out['average'] & emissions['poor'], fis3_env['poor'])
rule3fis3_env = ctrl.Rule(fis2_env_out['good'] & emissions['average'], fis3_env['average'])
rule4fis3_env = ctrl.Rule(fis2_env_out['poor'] & emissions['average'], fis3_env['poor'])
rule5fis3_env = ctrl.Rule(fis2_env_out['average'] & emissions['average'], fis3_env['average'])
rule6fis3_env = ctrl.Rule(fis2_env_out['good'] & emissions['average'], fis3_env['average'])
rule7fis3_env = ctrl.Rule(fis2_env_out['poor'] & emissions['good'], fis3_env['average'])
rule8fis3_env = ctrl.Rule(fis2_env_out['average'] & emissions['good'], fis3_env['average'])
rule9fis3_env = ctrl.Rule(fis2_env_out['good'] & emissions['good'], fis3_env['good'])

#FIS4_ENV = FIS1_ENV + FIS3_ENV
rule1fis4_env = ctrl.Rule(fis1_env_out['poor'] & fis3_env_out['poor'], fis4_env['poor'])
rule2fis4_env = ctrl.Rule(fis1_env_out['average'] & fis3_env_out['poor'], fis4_env['poor'])
rule3fis4_env = ctrl.Rule(fis1_env_out['good'] & fis3_env_out['average'], fis4_env['average'])
rule4fis4_env = ctrl.Rule(fis1_env_out['poor'] & fis3_env_out['average'], fis4_env['poor'])
rule5fis4_env = ctrl.Rule(fis1_env_out['average'] & fis3_env_out['average'], fis4_env['average'])
rule6fis4_env = ctrl.Rule(fis1_env_out['good'] & fis3_env_out['average'], fis4_env['average'])
rule7fis4_env = ctrl.Rule(fis1_env_out['poor'] & fis3_env_out['good'], fis4_env['average'])
rule8fis4_env = ctrl.Rule(fis1_env_out['average'] & fis3_env_out['good'], fis4_env['average'])
rule9fis4_env = ctrl.Rule(fis1_env_out['good'] & fis3_env_out['good'], fis4_env['good'])
```

```python
rule1fis4_env.view()
```

```python
fis1_env_ctrl = ctrl.ControlSystem([rule1fis1_env,rule2fis1_env,rule3fis1_env,rule4fis1_env,rule5fis1_env,rule6fis1_env,rule7fis1_env,rule8fis1_env,rule9fis1_env])
fis2_env_ctrl = ctrl.ControlSystem([rule1fis2_env,rule2fis2_env,rule3fis2_env,rule4fis2_env,rule5fis2_env,rule6fis2_env,rule7fis2_env,rule8fis2_env,rule9fis2_env])
fis3_env_ctrl = ctrl.ControlSystem([rule1fis3_env,rule2fis3_env,rule3fis3_env,rule4fis3_env,rule5fis3_env,rule6fis3_env,rule7fis3_env,rule8fis3_env,rule9fis3_env])
fis4_env_ctrl = ctrl.ControlSystem([rule1fis4_env,rule2fis4_env,rule3fis4_env,rule4fis4_env,rule5fis4_env,rule6fis4_env,rule7fis4_env,rule8fis4_env,rule9fis4_env])
```

```python
fissing1_env = ctrl.ControlSystemSimulation(fis1_env_ctrl)
fissing2_env = ctrl.ControlSystemSimulation(fis2_env_ctrl)
fissing3_env = ctrl.ControlSystemSimulation(fis3_env_ctrl)
fissing4_env = ctrl.ControlSystemSimulation(fis4_env_ctrl)
```

```python
#Input
fissing1_env.input['Material usage'] = 2
fissing1_env.input['Energy usage'] = 8
fissing2_env.input['Water usage'] = 2
fissing2_env.input['Waste'] = 10
#Crunch the numbers
fissing1_env.compute()
fissing2_env.compute()
fissing3_env.input['Fist 2 env out'] = fissing2_env.output['FIS2_ENV']
fissing3_env.input['Emissions'] = 5
fissing3_env.compute()
fissing4_env.input['Fist 1 env out'] = fissing1_env.output['FIS1_ENV']
fissing4_env.input['Fist 3 env out'] = fissing3_env.output['FIS3_ENV']
fissing4_env.compute()
```

```python
print(fissing1_env.output['FIS1_ENV'])
fis1_env.view(sim=fissing1_env)
```

```python
print(fissing2_env.output['FIS2_ENV'])
fis2_env.view(sim=fissing2_env)
```
